```python
salary_bins=[0, 20000, 40000, 60000, 80000, 100000, 150000, 200000, 500000, float('inf')]
salary_labels=['0-20k', '20-40k', '40-60k', '60-80k', '80-100k', '100-150k', '150-200k', '200-500k', '500k+']
```

```python
df['salary_range']=pd.cut(df['annual_salary'], bins=salary_bins, labels=salary_labels)
```

```python
df['salary_range'].value_counts().sort_index()
```

```python
df.shape
```

```python
# I used pd.cut() to bin salary data into ranges for analysis. Using the binning approach I identified outliers which I removed as they represented data errors 
df=df[(df['annual_salary']>=20000)&(df['annual_salary']<=500000)]
```

```python
df.shape
```

```python
df=df.drop(columns=['salary_range'])
```

```python
#Null values resulting in less than 2% of total data so I thought they are safe to remove. 
df = df.dropna(subset=['industry', 'education', 'gender'])
```

```python
#Here I realised I would have to do a total_compensation column to combine annual salary with bonus to get a realistic picture of total compensation 
df['extra_compensation'].isnull().sum()
```

```python
df['extra_compensation']=df['extra_compensation'].fillna(0)
```

```python
df['extra_compensation'].isnull().sum()
```

```python
df['total_compensation']=df['annual_salary']+df['extra_compensation']
```
